# Feature Engineering

```python
# features and Target Variables
X = df[['Ia','Ib', 'Ic', 'Va', 'Vb', 'Vc']] # Features
y = df['combined_faults_label'] # target
```

```python
# Removing negative values from the features
# All values are between 0 and about 2700

for col in X.columns:
  X[col] = X[col] + abs(X[col])

# X
```

```python
# Splitting data into train and test sets
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# # Standardizing the training dataset
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# # Initialize the scaler
scaler = MinMaxScaler()

# # Fit the scaler on the training data
scaler.fit(X_train)

# # Scale the training data
X_train_scaled = scaler.transform(X_train)

# # Scale the test data
X_test_scaled = scaler.transform(X_test)
```

#LOGISTIC REGRESSION

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Logistic Regression classifier
# The 'lbfgs' solver is a good default choice; for very large datasets or multiclass classification, 'sag' or 'saga' might be more appropriate
logreg = LogisticRegression(multi_class='ovr', solver='lbfgs', max_iter=1000)

# Fit the model to the training data
logreg.fit(X_train_scaled, y_train)

# Predict the labels for the test set
y_pred = logreg.predict(X_test_scaled)

# Predicting on the training set
train_predictions = logreg.predict(X_train_scaled)

# Calculating the accuracy on the training set
train_accuracy = accuracy_score(y_train, train_predictions)

print(f"Training Set Accuracy: {train_accuracy}")

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print(f"Test Set Accuracy: {accuracy:.2f}")

# Print a classification report to see metrics by class
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

# Define the model with the desired settings
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)

# Create a pipeline that first standardizes the data then fits the model
# This ensures each fold is scaled based on its training data
pipeline = make_pipeline(StandardScaler(), logreg)

# Choose the number of folds for cross-validation, 10 is a common choice
num_folds = 10

# Perform cross-validation
scores = cross_val_score(pipeline, X, y, cv=num_folds, scoring='accuracy')

# Calculate the average accuracy and the standard deviation
average_accuracy = scores.mean()
standard_deviation = scores.std()

# Output the results
print(f'Cross-validated accuracy: {average_accuracy:.3f}')
print(f'Standard Deviation: {standard_deviation:.3f}')
```

**Tuning the parameters for k-Nearest Neighbours**

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV, KFold
# StratifiedkFold retains the percentages of the each class in each fold
knn = KNeighborsClassifier()
folds = 10
kfold = StratifiedKFold(folds, shuffle=True, random_state=42)

# tuning the 'n_neighbors' parameter to find the best model

n_neighbors=[2,3,4,5,6,7,8,8,10,11,12]

#Using scaled X as k nearest neighbours is a distance based algorithm
knn_grid = GridSearchCV(knn ,{'n_neighbors': n_neighbors},
                        refit=True ,cv=kfold, scoring='accuracy')
knn_grid.fit(X_train_scaled, y_train)
knn_grid.best_params_

```

```python
# plotting the cross validation error
from sklearn.model_selection import cross_validate
scores = []

for n in n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=n)
    M_CV = cross_validate(knn, X_train_scaled, y_train, cv=kfold, scoring='accuracy')
    scores.append(1 - np.mean(M_CV['test_score']))

ax = plt.subplots(figsize=(10,6))[1]
ax.plot(n_neighbors, scores, color='purple')
ax.set_xlabel('Number of Neighbors')
ax.set_ylabel('Cross validation Error')
```

```python
#Decision trees

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(max_depth = 10, random_state=0)

# Train the classifier on the training data
dt_classifier.fit(X_train, y_train)

# Predicting on the training set
train_predictions = dt_classifier.predict(X_train)

# Calculating the accuracy on the training set
train_accuracy = accuracy_score(y_train, train_predictions)

print(f"Training Set Accuracy: {train_accuracy}")


# Predict the labels for the test set
y_pred = dt_classifier.predict(X_test)

# Evaluate the performance
print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# Create a pipeline that first standardizes the data then fits the model
# This ensures each fold is scaled based on its training data
pipeline = make_pipeline(dt_classifier)

# Choose the number of folds for cross-validation, 10 is a common choice
num_folds = 10

# Perform cross-validation
scores = cross_val_score(pipeline, X, y, cv=num_folds, scoring='accuracy')

# Calculate the average accuracy and the standard deviation
average_accuracy = scores.mean()
standard_deviation = scores.std()

# Output the results
print(f'Cross-validated accuracy: {average_accuracy:.3f}')
print(f'Standard Deviation: {standard_deviation:.3f}')
```

## Tuning the parameters of the Decision Tree

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV, KFold
# StratifiedkFold retains the percentages of the each class in each fold

folds = 10
kfold = StratifiedKFold(folds, shuffle=True, random_state=42)

# tuning the 'max_depth' parameter to find the best tree classifier using cross
# validation

max_depths = range(1,21)
tree = DecisionTreeClassifier()

dt_grid = GridSearchCV(tree ,{'max_depth': max_depths},
                        refit=True ,cv=kfold, scoring='accuracy')
dt_grid.fit(X_train, y_train)
dt_grid.best_params_
```

```python
# plot the cross validation error
from sklearn.model_selection import cross_validate
scores = []

for depth in max_depths:
    tree = DecisionTreeClassifier(max_depth=depth)
    M_CV = cross_validate(tree, X_train, y_train, cv=kfold, scoring='accuracy')
    scores.append(1 - np.mean(M_CV['test_score']))

ax = plt.subplots(figsize=(10,6))[1]
ax.plot(max_depths, scores)
ax.set_xlabel('Tree Depth')
ax.set_ylabel('Cross validation Error')
```

```python
unpruned_dt_test_acc = accuracy_score(y_test, dt_grid.predict(X_test))
unpruned_dt_test_acc
```

```python
# pruning the best tree
tree = DecisionTreeClassifier(max_depth = dt_grid.best_params_['max_depth'])
ccp_path = tree.cost_complexity_pruning_path(X_train , y_train)

pruned_dt_grid = GridSearchCV(tree ,{'ccp_alpha': ccp_path.ccp_alphas},
                        refit=True ,cv=kfold, scoring='accuracy')

pruned_dt_grid.fit(X_train, y_train)
pruned_dt_test_acc = accuracy_score(y_test, pruned_dt_grid.predict(X_test))
pruned_dt_test_acc
```

```python
# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay

matrix = confusion_matrix(y_test, pruned_dt_grid.predict(X_test))
disp = ConfusionMatrixDisplay(matrix, display_labels=pruned_dt_grid.classes_)
disp.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Initialize the SVC classifier with default hyperparameters for the baseline model
baseline_svc = SVC(random_state=0)

# Fit the classifier to the scaled training data
baseline_svc.fit(X_train_scaled, y_train)

# Predict and evaluate on the training set
train_predictions_baseline = baseline_svc.predict(X_train_scaled)
train_accuracy_baseline = accuracy_score(y_train, train_predictions_baseline)
print(f"Baseline Training Set Accuracy: {train_accuracy_baseline}")

# Predict and evaluate on the test set
y_pred_svc_baseline = baseline_svc.predict(X_test_scaled)
print(classification_report(y_test, y_pred_svc_baseline))
accuracy_svc_baseline = accuracy_score(y_test, y_pred_svc_baseline)
print(f'Baseline Test Accuracy: {accuracy_svc_baseline}')
```

```python
 # Define the parameter range for GridSearch
param_grid = {
    'C': [0.1, 1, 10, 100],  # Example values
    'gamma': [1, 0.1, 0.01, 0.001],  # Example values
    'kernel': ['rbf']  # Focusing on 'rbf' kernel for simplicity
}

# Initialize the GridSearchCV object with the SVC classifier and the parameter grid
grid = GridSearchCV(SVC(random_state=0), param_grid, refit=True, verbose=3, cv=5)

# Fit the grid search to the data
grid.fit(X_train_scaled, y_train)

# Print the best parameters found by GridSearchCV
print("Best parameters found by GridSearchCV:")
print(grid.best_params_)

# Predict and evaluate on the training set
train_predictions = grid.predict(X_train_scaled)
train_accuracy = accuracy_score(y_train, train_predictions)
print(f"Training Set Accuracy: {train_accuracy}")

# Predict and evaluate on the test set
y_pred_svc = grid.predict(X_test_scaled)
print(classification_report(y_test, y_pred_svc))
accuracy_svc = accuracy_score(y_test, y_pred_svc)
print(f'Test Accuracy: {accuracy_svc}')
```

```python
# 'grid' is the GridSearchCV object after fitting it to the training data
best_svc_model = grid.best_estimator_

# use this model to make predictions or further evaluate its performance
#
test_predictions = best_svc_model.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test, test_predictions)

# Print out the performance
print(f'Test Accuracy of the best SVC model: {test_accuracy}')
```

```python
# Visualization
cv_results = grid.cv_results_
scores = cv_results['mean_test_score']
C_values = param_grid['C']
gamma_values = param_grid['gamma']

# Reshaping the scores array
scores_matrix = np.array(scores).reshape(len(C_values), len(gamma_values))

# Plotting
plt.figure(figsize=(8, 6))
for idx, val in enumerate(C_values):
    plt.plot(gamma_values, scores_matrix[idx, :], label=f'C={val}')

plt.legend()
plt.xlabel('Gamma')
plt.ylabel('Mean Test Score')
plt.title('SVC Performance for Different C and Gamma Values')
plt.show()
```

```python
# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix


matrix = confusion_matrix(y_test, best_svc_model.predict(X_test_scaled))
disp = ConfusionMatrixDisplay(matrix, display_labels=best_svc_model.classes_)
disp.plot()
```
